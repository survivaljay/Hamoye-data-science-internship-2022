### Q 17. The Element with the highest sum in 2018

```python
a = pd.DataFrame(data.groupby('Element').sum()['Y2018'])

a.sort_values(by=['Y2018'], inplace=True, ascending=False)

highest_sum = a.head(5)

highest_sum
```
